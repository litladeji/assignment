# Iris Dataset Analysis

## Prepare the dataset

Use `numpy` or to load `data/iris.data`. Remember to look at the file contents to understand the data format.

```python
import numpy as np
import matplotlib.pyplot as plt
data_set = np.loadtxt("iris.data", delimiter=",", usecols=(0, 1, 2, 3))
# data_set = np.genfromtxt("iris.data", delimiter=",", usecols=(0, 1, 2, 3))
```

Create a map that returns the attribute name given the attribute index.

- 0 -> sepal length
- 1 -> sepal width
- 2 -> petal length
- 3 -> petal width

```python
attribute_map = {
    0: "sepal length",
    1: "sepal width",
    2: "petal length",
    3: "petal width"
}
```

## Data analysis

Write a function that prints the mean and another that prints the standard deviation of each attribute.

```python
def print_means(data):
    means = np.mean(data, axis=0)
    means_1 = []
    # print(means)
    for i,value in enumerate(means):
       means_1.append(f"{attribute_map[i]}:{value:.3f}") 
    print(means_1)
    

def print_std_devs(data):
    stds = np.std(data, axis=0)
    # print(stds)
    stds_1 = []
    # print(means)
    for i,value in enumerate(stds):
       stds_1.append(f"{attribute_map[i]}:{value:.3f}") 
    print(stds_1)
```

Use the functions you defined above to print the mean of the attributes for each class.

```python
print_means(data_set)
print_std_devs(data_set)
```

Now, let's plot this data. Use `matplotlib` to create scatter plots for petal and sepal. Use different colors for each class.

```python
full_data = np.genfromtxt("iris.data", delimiter=",", dtype=str)
data = full_data[:, :4].astype(float)
labels = full_data[:, 4]

class_colors = {'Iris-setosa': 'red', 'Iris-versicolor': 'green', 'Iris-virginica': 'blue'}

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

for class_name in np.unique(labels):
    class_data = data[labels == class_name]
    ax1.scatter(class_data[:, 0], class_data[:, 1], label=class_name, color=class_colors[class_name])

ax1.set_xlabel("Sepal Length")
ax1.set_ylabel("Sepal Width") 
ax1.set_title("Sepal Length vs Width")
ax1.grid(True)
ax1.legend()

for class_name in np.unique(labels):
    class_data = data[labels == class_name]
    ax2.scatter(class_data[:, 2], class_data[:, 3], label=class_name, color=class_colors[class_name])

ax2.set_xlabel("Petal Length")
ax2.set_ylabel("Petal Width")
ax2.set_title("Petal Length vs Width")
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```
